```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python

url="https://raw.githubusercontent.com/sofinu/data-sharks/main/world_population.csv"
df = pd.read_csv(url)
num_filas = len(df)

num_columnas = len(df.columns)
print("Número de filas:", num_filas)
print("Número de columnas:", num_columnas)

df
```

```python
#Especificar las dos columnas de interés
pais='Country/Territory'
poblacion='1970 Population'

#Crear un DataFrame con las dos columnas
datos=df[[pais, poblacion]]

#Obtener las filas con los valores más altos en la columna precio
top_20_paises_menos_poblados=datos.nsmallest(20, poblacion)

print(top_20_paises_menos_poblados)
```

```python
import matplotlib.pyplot as plt

# Suponiendo que 'df' es tu DataFrame y contiene los datos
# de los 20 países menos poblados en 1970

# Columnas a utilizar
top_20_paises_menos_poblados = df.nsmallest(20, '1970 Population')

# Crear un gráfico de barras
plt.bar(top_20_paises_menos_poblados['Country/Territory'], top_20_paises_menos_poblados['1970 Population'], edgecolor='blue', color='blue')

# ejes X Y
plt.xlabel('Country/Territory')
plt.ylabel('1970 Population')

# Configurar el espacio entre las etiquetas del eje x
plt.xticks(rotation=45, ha='right', va='top')  # Alinear las etiquetas a la derecha y ajustar la alineación vertical

plt.title("Top 20 países menos poblados a nivel mundial en 1970")
plt.tight_layout()  # Ajustar el diseño para evitar recortes
plt.show()
```
